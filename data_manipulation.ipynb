```python
%sql
SELECT DISTINCT
  Year,
  Quarter,
  OriginAirportID,
  Origin,
  OriginState,
  DestAirportID,
  Dest,
  DestState, 
  replace(AirportGroup, ":", " -> ") as Route,
  CASE 
    WHEN length(AirportGroup) > 7 THEN 0 
    ELSE 1
  END as NonStop,
  MIN(MktFare) over (partition by Year, Quarter, OriginAirportID, Origin, OriginState, DestAirportID, Dest, DestState) as Fare_min,
  MAX(MktFare) over (partition by Year, Quarter, OriginAirportID, Origin, OriginState, DestAirportID, Dest, DestState) as Fare_max,
  AVG(MktFare) over (partition by Year, Quarter, OriginAirportID, Origin, OriginState, DestAirportID, Dest, DestState) as Fare_mean,
  percentile_cont(0.25) within group (order by MktFare) over (partition by Year, Quarter, OriginAirportID, Origin, OriginState, DestAirportID, Dest, DestState) as Fare_25_pct,
  percentile_cont(0.5) within group (order by MktFare) over (partition by Year, Quarter, OriginAirportID, Origin, OriginState, DestAirportID, Dest, DestState) as Fare_median,
  percentile_cont(0.75) within group (order by MktFare) over (partition by Year, Quarter, OriginAirportID, Origin, OriginState, DestAirportID, Dest, DestState) as Fare_75_pct,
  SUM(Passengers) over (partition by Year, Quarter, OriginAirportID, Origin, OriginState, DestAirportID, Dest, DestState) as Passengers
FROM airline_origin_dest_data.airline_data_table
```
